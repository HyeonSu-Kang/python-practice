반복문 5 - zip

```python
for z in zip(a,b):
  print(z)
  print(z[0]+z[1])
```

```python
for idx, z in enumerate(zip(a,b)):
  print("인덱스{}: a={},b={}" .format(idx,z[0],z[1]))
```

pandas.DataFrame 만들기 (리스트 활용)


```python
df=pd.DataFrame([[1,2,'C'],
                 [4,5,'B'],
                 [7,8,'B']],
                index=['stu1','stu2','stu3'],
                columns=['math','sci','Grade'])
df
```

pandas.DataFrame 에서 특정 열 추출하기기


```python
x=df['math']
print(x)
print(type(x),'\n')

x_list=df['math'].to_list()
print(x_list)
print(type(x_list))
```

그래프 그리기기

```python
import matplotlib.pyplot as plt
%matplotlib inline

x_math=df['math'].to_list()
x_sci=df['sci'].to_list()

plt.figure(figsize=(5,5))
plt.scatter(x_math,x_sci)
plt.show()
```

그래프 그리기 - 딕셔너리 활용하기기

```python
col_dict={'A':'red','B':'green','C':'orange'}
x_grd=df['Grade'].to_list()

plt.figure(figsize=(5,5))
for z in zip(x_math,x_sci,x_grd):
  plt.scatter(z[0],z[1],c=col_dict[z[2]])
plt.show()
```
